```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('creditcard.csv')
```

```python
sns.countplot(data = df, x = 'Class')
plt.title("distribution of data (froud = 1,genuine = 0)")
plt.show()          
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Amount'],bins = 20,kde = True)
plt.title('transaction of amount')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Time"], bins = 20, kde = True)
plt.title('transaction time distribution')
plt.show()
             
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap='coolwarm_r', annot=False)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
fraud_percentage = (df['Class'].value_counts()[1] / df['Class'].value_counts().sum()) * 100
print(f"Fraudulent Transactions: {fraud_percentage:.4f}%")
```
